```python
from sklearn import datasets
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from sklearn.cluster import KMeans


from sklearn.datasets import fetch_openml

```

```python
import pandas as pd
from markdown import markdown
from IPython.core.display import HTML
```

```python
X, _ = fetch_openml('mnist_784', version=1, return_X_y=True)
```

```python
X= X/X.max()
```

## Steps 

- Feature engineering
- Train/Validation/Test splits
- Fit the model 
- Evaluate the model 

```python
model = KMeans(n_clusters = 10,max_iter=1000)

model.fit_transform(X[:1000])
```

```python
labels = model.labels_ 
```

```python
group_0 = np.where( labels==4)[0]
group_0[0:10]
```

```python
for i in range(5):
    example = X[group_0[i]].reshape(28,28)
    plt.figure( figsize=(5,5))
    plt.imshow( example , cmap='gray', vmin=0) 
    plt.show()
```
